```python
import numpy as np
```

```python
arr = np.array([[4,6,8,1],[5,1,8,2],[9,5,2,1],[6,8,0,1]])
print(arr)
```

```python
arr2 = np.array([10,20,30,40])
print(arr2)

arr = arr + arr2

arr3 = arr2.reshape(-1,1)

print(arr * arr3)
```
