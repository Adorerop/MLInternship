```python
import pandas as pd
df = pd.read_csv('Car_Details.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
#Subtract year from 2023 and reassign it in same column
df['year']=2023 - df['year']
```

```python
df.head()
```

```python
# Standardscalling on year and km_driven
from sklearn.preprocessing import StandardScaler

standard1 = StandardScaler()
df[['year']] = standard1.fit_transform(df[['year']])

standard2 = StandardScaler()
df[['km_driven']] = standard2.fit_transform(df[['km_driven']])
```

```python
df.head()
```

```python
#Onehot encoding on fuel column, convert onehot encoded array to2d dataframe, Merge 2 dataframes.
from sklearn.preprocessing import MinMaxScaler,StandardScaler,OneHotEncoder
ohe = OneHotEncoder(sparse=False)
result = ohe.fit_transform(df[['fuel']])
new_df = pd.DataFrame(result, columns=ohe.categories_[0])
df = pd.concat([df,new_df],axis = 1)
del df['fuel']
```

```python
#Onehot encoding on seller_type	transmission	owner
```

```python
ohe = OneHotEncoder(sparse=False)
result = ohe.fit_transform(df[['seller_type']])
new_df = pd.DataFrame(result, columns=ohe.categories_[0])
df = pd.concat([df,new_df],axis = 1)
del df['seller_type']
```

```python
ohe = OneHotEncoder(sparse=False)
result = ohe.fit_transform(df[['transmission']])
new_df = pd.DataFrame(result, columns=ohe.categories_[0])
df = pd.concat([df,new_df],axis = 1)
del df['transmission']
```

```python
ohe = OneHotEncoder(sparse=False)
result = ohe.fit_transform(df[['owner']])
new_df = pd.DataFrame(result, columns=ohe.categories_[0])
df = pd.concat([df,new_df],axis = 1)
del df['owner']
```

```python
df.head()
```

```python
# Write function that return first word from string
Input = "I am from India"
s = Input.split()
print(s[0].lower())
```

```python
def get_company(s):
  s = s.split()
  first = s[0].lower()
  return(first)
```

```python
get_company(Input)
```

```python
df['name'] = df['name'].apply(get_company)
```

```python
df.head()
```

```python
df['name'].unique()
```

```python
ohe = OneHotEncoder(sparse=False)
result = ohe.fit_transform(df[['name']])
new_df = pd.DataFrame(result, columns=ohe.categories_[0])
df = pd.concat([df,new_df],axis = 1)
del df['name']
```

```python
df.head()
```

```python
Y = df['selling_price']
```

```python
del df['selling_price']
```

```python
X = df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

#LinearRegression

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python

model = LinearRegression()
model.fit(X_train,y_train)

pred = model.predict(X_test)

mse = mean_squared_error(y_test, pred)
print(mse)
```

#DecisionTreeRegresson


```python
from sklearn.tree import DecisionTreeRegressor
```

```python

model = DecisionTreeRegressor()
model.fit(X_train,y_train)

pred = model.predict(X_test)

mse = mean_squared_error(y_test, pred)
print(mse)
```

```python
import warnings
warnings.filterwarnings("ignore")


for i in range(3,21):
    for j in range(3,9):
        clf = DecisionTreeRegressor(max_depth=i,min_samples_leaf=j)
        clf.fit(X_train, y_train)
        pred=clf.predict(X_test)
        
        mse=mean_squared_error(y_test,pred)
        print(mse,i,j)
```

#RandomForestRegressor

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html#:~:text=A%20random%20forest%20regressor.,accuracy%20and%20control%20over%2Dfitting


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X_train,y_train)

pred = model.predict(X_test)

mse = mean_squared_error(y_test, pred)
print(mse)
```
